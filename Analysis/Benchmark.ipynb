```python
# Reading in the data:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.interpolate import interp1d
trainData = pd.read_csv('../data/1e7data_train.csv', index_col=0)
testData = pd.read_csv('../data/1e7data_holdout.csv', index_col=0)
df = trainData
dftest = testData
df.tail()
```

```python
df["sumRH"] = df["mindaOverRH"] + df["maxdaOverRH"]
dftest["sumRH"] = dftest["mindaOverRH"] + dftest["maxdaOverRH"]
```

```python
pred = dftest["sumRH"] > 14.
pred[:10]
```

```python
from sklearn import metrics
for thresh in np.linspace(12,18,30):
    pred = df["sumRH"] > thresh
    print("thresh = {0:.3}, acc = {1:.3}, prec = {2:.3}, rec = {3:.3}".format(thresh, metrics.accuracy_score(df['Stable'].values, pred), metrics.precision_score(df['Stable'].values, pred), metrics.recall_score(df['Stable'].values, pred)))
```
