```python
# Reading in the data:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.interpolate import interp1d
trainData = pd.read_csv('../data/1e7data_train.csv', index_col=0)
testData = pd.read_csv('../data/1e7data_holdout.csv', index_col=0)
df = trainData
dftest = testData
df.tail()
```

```python
df["sumRH"] = df["mindaOverRH"] + df["maxdaOverRH"]
dftest["sumRH"] = dftest["mindaOverRH"] + dftest["maxdaOverRH"]
```

```python
pred = dftest["sumRH"] > 14.
pred[:10]
```

```python
from sklearn import metrics
Nvalues = 200
precision, recall, thresholds = np.zeros(Nvalues), np.zeros(Nvalues), np.zeros(Nvalues)
for i, thresh in enumerate(np.linspace(10.,18.,Nvalues)):
    pred = df["sumRH"] > thresh
    thresholds[i] = thresh
    precision[i] = metrics.precision_score(df['Stable'].values, pred)
    recall[i] = metrics.recall_score(df['Stable'].values, pred)
    print("thresh = {0:.6}, acc = {1:.6}, prec = {2:.6}, rec = {3:.6}".format(thresholds[i], metrics.accuracy_score(df['Stable'].values, pred), precision[i], recall[i]))
```

```python
np.savetxt('../data/thresholdsLM.txt', thresholds)
np.savetxt('../data/recallLM.txt', recall)
np.savetxt('../data/precisionLM.txt', precision)
```
