```python
import rebound
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

masterdf = pd.read_csv('../data/1e7datasetcorr.csv', index_col=0)

fig, axarr = plt.subplots(figsize=(15,10), nrows=2)
for sim_id in range(5000):
    columns = ['Time']
    for i in ['1', '2', '3']:
        columns += ['a'+i, 'P'+i, 'e'+i, 'pomega'+i, 'inc'+i, 'Omega'+i, 'f'+i]

    filename = '/mnt/raid-cita/dtamayo/Stability/data/corrdata/output'+str(sim_id)+'.csv'
    df = pd.read_csv(filename, names=columns)
    
    color = 'r'
    if masterdf.loc[sim_id, 'Stable'] == True:
        color = 'b'
        
    a0 = df['a1'].mean()
    e0 = df['e1'].mean()
    axarr[0].plot(df['Time'], df['a1']/a0, '.', color=color, alpha=0.1)
    axarr[1].plot(df['Time'], df['e1'], '.', color=color, alpha=0.1)
    
```
